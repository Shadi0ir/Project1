```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
%matplotlib inline
from matplotlib.pyplot import figure
from pprint import pprint
```

```python
# Import our data into pandas from CSV
Data_source = '../Shadi-Charts/11_country.csv'
file_df = pd.read_csv(Data_source)

file_df
```

```python
#Clean and re-arrange the columns
data_df=file_df[['Country','Danceability','Loudness','Valence','Speechiness','Acousticness','Instramentalness','Tempo','Duration (ms)']]
# clean a colum name
data_df=data_df.rename(columns={"Duration (ms)":"Duration"})

data_df
```

```python
# Calculate the mean for all musical features per country

pd.options.display.float_format = '{:,.2f}'.format
#Group by country and calculate mean

Danceability_mean=data_df.groupby(['Country']).mean()
Danceability_mean
```

```python
#retrive columns
data_df_Danceability=data_df[["Country","Danceability"]]
#Group by country and calculate mean

Danceability_mean=data_df_Danceability.groupby(['Country']).mean()
Danceability_mean
```

```python
# Comparing Mean for diffirent musica features between countries

# Danceability
# Create a bar chart based off of the mean by country
plt.figure(figsize=(3.841,10),dpi=100)
Danceability_mean_chart = Danceability_mean.plot(kind='bar',color=('orange'))
plt.ylim(0.3, 0.8)
# Set the xlabel and ylabel using class methods
Danceability_mean_chart.set_xlabel("Country")
Danceability_mean_chart.set_ylabel("Average Danceability")

plt.tight_layout()


plt.savefig("Danceability", dpi=1000)
plt.show()


# Loudness
#retrive columns
plt.figure(figsize=(19.20,10.80))
data_df_Loudness=data_df[["Country","Loudness"]]
#Group by country and calculate mean

Loudness_mean=data_df_Loudness.groupby(['Country']).mean()
Loudness_mean


# Create a bar chart based off of the mean by country
Loudness_mean_chart = Loudness_mean.plot(kind='bar')
# Set the xlabel and ylabel using class methods
Loudness_mean_chart.set_xlabel("Country")
Loudness_mean_chart.set_ylabel("Average Loudness")


plt.show()
plt.tight_layout()

# Acousticness
#retrive columns
plt.figure(figsize=(19.20,10.80))
data_df_Acousticness=data_df[["Country","Acousticness"]]
#Group by country and calculate mean

Acousticness_mean=data_df_Acousticness.groupby(['Country']).mean()
Acousticness_mean


# Create a bar chart based off of the mean by country
Acousticness_mean_chart = Acousticness_mean.plot(kind='bar',color=('g'))
# Set the xlabel and ylabel using class methods
Acousticness_mean_chart.set_xlabel("Country")
Acousticness_mean_chart.set_ylabel("Average Acousticness")


plt.show()
plt.tight_layout()

# Tempo
#retrive columns
plt.figure(figsize=(19.20,10.80))
data_df_Tempo=data_df[["Country","Tempo"]]
#Group by country and calculate mean

Tempo_mean=data_df_Tempo.groupby(['Country']).mean()
Tempo_mean


# Create a bar chart based off of the mean by country
Tempo_mean_chart = Tempo_mean.plot(kind='bar',color=('yellow'))
# Set the xlabel and ylabel using class methods
Tempo_mean_chart.set_xlabel("Country")
Tempo_mean_chart.set_ylabel("Average Tempo")


plt.show()
plt.tight_layout()


#Duration
#retrive columns
plt.figure(figsize=(19.20,10.80))
data_df_Duration=data_df[["Country","Duration"]]
#Group by country and calculate mean

Duration_mean=data_df_Duration.groupby(['Country']).mean()
Duration_mean


# Create a bar chart based off of the mean by country
Duration_mean_chart = Duration_mean.plot(kind='bar',color=('red'))
# Set the xlabel and ylabel using class methods
Duration_mean_chart.set_xlabel("Country")
Duration_mean_chart.set_ylabel("Average Duration")


plt.show()
plt.tight_layout()

#Valence
#retrive columns
plt.figure(figsize=(19.20,10.80))
data_df_Valence=data_df[["Country","Valence"]]
#Group by country and calculate meanValence

Valence_mean=data_df_Valence.groupby(['Country']).mean()
Valence_mean


# Create a bar chart based off of the mean by country
Valence_mean_chart = Valence_mean.plot(kind='bar',color=('purple'))
# Set the xlabel and ylabel using class methods
Valence_mean_chart.set_xlabel("Country")
Valence_mean_chart.set_ylabel("Average Duration")


plt.show()
plt.tight_layout()
```
